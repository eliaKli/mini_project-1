## Laptop Price Dataset


```python
# Import libraries

import pandas as pd       # לעבודה עם DataFrame
import numpy as np        # פונקציות מתמטיות וסטטיסטיות
import matplotlib.pyplot as plt  # ליצירת גרפים
import seaborn as sns     # לויזואליזציות מתקדמות

# הגדרות כלליות לגרפים
sns.set_theme(style="whitegrid")   # סגנון גריד נקי
plt.rcParams['figure.figsize'] = (10,6)  # גודל גרפים

# Load the dataset

# החליפי את הנתיב למיקום הקובץ אצלך
df = pd.read_csv(r"C:\Users\elia1\.vscode\PyProjects\GitHub Projects\mini_project\laptop_price - dataset.csv")

# Initial exploration of the data

print("Head of the dataset:")
display(df.head())

# מידע כללי על העמודות, סוגי הנתונים וחסרונות
print("\nDataset info:")
df.info()

# סטטיסטיקות בסיסיות למשתנים מספריים
print("\nSummary statistics:")
display(df.describe())
```

Task 1: Visualizing Laptop Prices
In this section, we visualize the distribution of laptop prices across the entire dataset. I will use a Histogram with a KDE (Kernel Density Estimate) line to see the frequency of different price points and identify the general price range of most laptops

```python
# יצירת דף לגרף בגודל מתאים
plt.figure(figsize=(10, 6))

# יצירת היסטוגרמה למחירי הלפטופים
sns.histplot(df['Price (Euro)'], kde=True, color='teal')

# הוספת כותרות (Axes titles and plot title)
plt.title('Distribution of Laptop Prices', fontsize=14)
plt.xlabel('Price (Euro)', fontsize=12)
plt.ylabel('Number of Laptops', fontsize=12)

# הצגת הגרף
plt.show()
```

Observation: The histogram shows that the majority of laptops are priced in the lower-to-mid range (around 300 to 1200), with a few expensive laptops acting as potential outliers on the right side of the graph.

Task 2: Average Price by Company
In this section, I will group the data by the 'Company' column and calculate the mean price for each. This will help us identify which brand is the most expensive on average and how prices vary across different manufacturers.

```python
# חישוב המחיר הממוצע לכל חברה (סידור מהגבוה לנמוך)
avg_price_company = df.groupby('Company')['Price (Euro)'].mean().sort_values(ascending=False)

# הצגת הנתונים כטקסט (כדי לענות על השאלה "מה המחיר הממוצע")
print("Average laptop price in euros for each company:")
print(avg_price_company)

# מציאת החברה היקרה ביותר
most_expensive = avg_price_company.idxmax()
highest_avg = avg_price_company.max()
print(f"\nThe company with the most expensive laptops on average is: {most_expensive} ({highest_avg:.2f} Euro)")

# יצירת גרף עמודות להמחשה ויזואלית
plt.figure(figsize=(12, 6))
avg_price_company.plot(kind='bar', color='skyblue', edgecolor='black')

# כותרות ושמות לצירים (חובה לפי ההנחיות)
plt.title('Average Laptop Price per Company', fontsize=14)
plt.xlabel('Company Name', fontsize=12)
plt.ylabel('Average Price (Euro)', fontsize=12)
plt.xticks(rotation=45) # סיבוב שמות החברות כדי שיהיה קל לקרוא

plt.show()
```

Conclusion: Based on the analysis, Razer has the highest average laptop price. This indicates that their product line is positioned in the premium market segment.
