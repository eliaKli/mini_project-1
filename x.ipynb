Task 1: Visualizing Laptop Prices
In this section, we visualize the distribution of laptop prices across the entire dataset. I will use a Histogram with a KDE (Kernel Density Estimate) line to see the frequency of different price points and identify the general price range of most laptops

```python
# יצירת דף לגרף בגודל מתאים
plt.figure(figsize=(10, 6))

# יצירת היסטוגרמה למחירי הלפטופים
sns.histplot(df['Price (Euro)'], kde=True, color='teal')

# הוספת כותרות (Axes titles and plot title)
plt.title('Distribution of Laptop Prices', fontsize=14)
plt.xlabel('Price (Euro)', fontsize=12)
plt.ylabel('Number of Laptops', fontsize=12)

# הצגת הגרף
plt.show()
```

Observation: The histogram shows that the majority of laptops are priced in the lower-to-mid range (around 300 to 1200), with a few expensive laptops acting as potential outliers on the right side of the graph.
