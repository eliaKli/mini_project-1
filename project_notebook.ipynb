## Laptop Prices by Company


In this section, I will group the data by the 'Company' column and calculate the mean price for each. This will help identify which brand is the most expensive on average and how prices vary across different manufacturers.

```python
# חישוב המחיר הממוצע לכל חברה (סידור מהגבוה לנמוך)
avg_price_company = df.groupby('Company')['Price (Euro)'].mean().sort_values(ascending=False)

# הצגת הנתונים כטקסט (כדי לענות על השאלה "מה המחיר הממוצע")
print("Average laptop price in euros for each company:")
print(avg_price_company)

# מציאת החברה היקרה ביותר
most_expensive = avg_price_company.idxmax()
highest_avg = avg_price_company.max()
print(f"\nThe company with the most expensive laptops on average is: {most_expensive} ({highest_avg:.2f} Euro)")

# יצירת גרף עמודות להמחשה ויזואלית
plt.figure(figsize=(12, 6))
avg_price_company.plot(kind='bar', color='skyblue', edgecolor='black')

# כותרות ושמות לצירים (חובה לפי ההנחיות)
plt.title('Average Laptop Price per Company', fontsize=14)
plt.xlabel('Company Name', fontsize=12)
plt.ylabel('Average Price (Euro)', fontsize=12)
plt.xticks(rotation=45) # סיבוב שמות החברות כדי שיהיה קל לקרוא

plt.show()
```

Based on the analysis, Razer has the highest average laptop price. This indicates that their product line is positioned in the premium market segment.

## Operating Systems Analysis


### Unique operating systems before cleaning


```python
# הצגת כל מערכות ההפעלה הייחודיות כפי שהן מופיעות בדאטה
df['OpSys'].unique()
# ספירת הופעות של כל מערכת הפעלה
df['OpSys'].value_counts()
```

### Cleaning and Unifying Operating System Categories


```python
# יצירת עותק של העמודה לצורך ניקוי
df['OpSys_clean'] = df['OpSys']

# איחוד כל גרסאות Windows
df.loc[df['OpSys_clean'].str.contains('Windows', case=False, na=False), 'OpSys_clean'] = 'Windows'

# איחוד כל גרסאות macOS
df.loc[df['OpSys_clean'].str.contains('Mac', case=False, na=False), 'OpSys_clean'] = 'macOS'

# איחוד מערכות Linux
df.loc[df['OpSys_clean'].str.contains('Linux', case=False, na=False), 'OpSys_clean'] = 'Linux'

# איחוד מחשבים ללא מערכת הפעלה
df.loc[df['OpSys_clean'].str.contains('No OS', case=False, na=False), 'OpSys_clean'] = 'No OS'
```

In this step, operating system names that refer to the same system but are written differently
(e.g. different versions of Windows) are unified into consistent categories.
This allows clearer comparison and visualization in later analysis.


### Unique Operating Systems After Cleaning


```python
df['OpSys_clean'].value_counts()
```

### Price Distribution by Operating System


```python
# קבלת רשימת מערכות ההפעלה הייחודיות לאחר ניקוי
operating_systems = df['OpSys_clean'].unique()

# יצירת גרף התפלגות מחירים לכל מערכת הפעלה
for os in operating_systems:
    subset = df[df['OpSys_clean'] == os]
    
    plt.figure()
    plt.hist(subset['Price (Euro)'], bins=30)
    plt.title(f'Price Distribution for {os}')
    plt.xlabel('Price (Euro)')
    plt.ylabel('Number of Laptops')
    plt.show()
```

## Relationship Between RAM and Laptop Price


### RAM Values Overview


```python
df['RAM (GB)'].value_counts().sort_index()
```

### Relationship Between RAM and Price


To analyze the relationship, I will use a scatter plot with a Regression Line. This line helps visualize the correlation between RAM size and price.

```python
# יצירת גרף פיזור עם קו מגמה
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='RAM (GB)', y='Price (Euro)', 
            scatter_kws={'alpha':0.5}, 
            line_kws={'color':'red'})

plt.title('Relationship Between RAM and Laptop Price', fontsize=14)
plt.xlabel('RAM (GB)', fontsize=12)
plt.ylabel('Price (Euro)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

An outlier is defined as a data point with an unusually high or low price relative to the majority of laptops with similar RAM values. 
Such points may represent premium models or data anomalies and can strongly influence visual interpretation and statistical measures.


### Outlier Detection Using Boxplots


```python
plt.figure()
sns.boxplot(data=df, x='RAM (GB)', y='Price (Euro)')
plt.title('Price Distribution by RAM')
plt.xlabel('RAM (GB)')
plt.ylabel('Price (Euro)')
plt.show()
```

The results suggest a positive relationship between RAM size and laptop price, although several high-priced outliers are present, particularly for higher RAM configurations.

### Extracting Storage Type from Memory


Feature Engineering - Storage Type Extraction.
In this step, I extract the storage technology (SSD, HDD, Flash, etc.) from the complex 'Memory' string. This allows us to categorize laptops by their storage type regardless of the capacity.

```python
def extract_storage_type(memory):
    if pd.isna(memory):
        return None

    memory = memory.upper()

    if 'SSD' in memory and 'HDD' in memory:
        return 'Multiple'
    elif 'FLASH' in memory:
        return 'Flash Storage'
    elif 'SSD' in memory:
        return 'SSD'
    elif 'HDD' in memory:
        return 'HDD'
    elif 'HYBRID' in memory:
        return 'Hybrid'
    else:
        return 'Other'

df['Storage type'] = df['Memory'].apply(extract_storage_type)

#בדיקה חוזרת
df[['Memory', 'Storage type']].head(10)

```
